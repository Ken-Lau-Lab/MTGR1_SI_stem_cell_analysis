## annotation for the brush border genes

```python
data = pd.read_csv("./brush_border_gene_exp_df.csv", header = 0, index_col= 0 )
```

```python
data.head()
```

```python
data.columns
```

```python
#new_brush_genes = data.columns[0:-1]
new_brush_genes = ['Actb', 'Plec', 'Vil1', 'Pls1', 'Ezr', 'Krt20', 'Dsp', 'Sptan1',
       'Actn4', 'Sptbn1', 'Cltc', 'Ctnna1',  'Eps8', 'Actr3', 'Flnb',
       'Cobl', 'Tprn', 'Cap1', 'Baiap2l1', 'Cdhr2', 'Cdhr5']
len( new_brush_genes)
```

```python
pastel = sn.color_palette("pastel", 8).as_hex()
pastel
```

```python
pastel[1]
```

```python
# abs_treatment_color = ['#a1c9f4', '#ffb482']
abs_treatment_color = ['#ffb482', '#a1c9f4']
```

```python
df = data
x = "condition"
y = "Actb"
order = ['WT', 'Mtgr1 -/-']

f,ax2 = plt.subplots( figsize = [5,5])
#sn.set(font_scale=1)


# ax2 = sn.stripplot( data = df,  x = x, y = y, 
#               palette = ['b', 'brown'], hue = 'condition', size = 1.5, alpha = 0.5)
sn.violinplot(data = df,  x = x, y = y, 
               palette=abs_treatment_color, hue = 'condition',  ax = ax2, kwargs={'alpha':'0.6', 'legend':False})
sn.stripplot( data = df,  x = x, y = y, 
             dodge = True,
               palette = ['chocolate', 'b'], hue = 'condition', size = 1.5, alpha = 0.5, ax = ax2)
ax2.get_legend().remove()
#ax1.set_ylabel(y)

pairs=[('WT', 'Mtgr1 -/-')]
annotator = Annotator(ax2, pairs, data=df, x=x, y=y, order=order)
annotator.configure(test='Mann-Whitney', text_format='star', loc='inside')
annotator.apply_and_annotate()
```

```python
# f,axs = plt.subplots( nrows=6, ncols=4, figsize = [20, 26])
# sn.set(font_scale=1)

# axs = axs.ravel()
# for i in range( len( new_brush_genes) ):
#     c = new_brush_genes[i]
    
#     sn.violinplot( data = df2,  x = 'condition', y = c, 
#                palette=list( ab.uns['treatment_colors']), hue = 'condition',
#               ax = axs[i], alpha = 0.6)
#     sn.stripplot( data = df2,  x = 'condition', y = c, 
#               palette = ['b', 'brown'], hue = 'condition',
#               ax = axs[i], size = 1.5, alpha = 0.5)
#     axs[i].set( ylabel= c)
```

```python
f,axs = plt.subplots( nrows=7, ncols=3, figsize = [18, 42])
sn.set(font_scale=1)

axs = axs.ravel()
for i in range( len( new_brush_genes) ):
    x = 'condition'
    y = new_brush_genes[i]
    sn.violinplot(data = data,  x = x, y = y, 
               palette=abs_treatment_color, hue = 'condition',  ax = axs[i], )
    sn.stripplot( data = data,  x = x, y = y, 
                 dodge = True,
                   palette = ['chocolate', 'b'], hue = 'condition', size = 1.5, alpha = 0.5, ax = axs[i])
    axs[i].get_legend().remove()
    #ax1.set_ylabel(y)

    pairs=[('WT', 'Mtgr1 -/-')]
    annotator = Annotator(axs[i], pairs, data=data, x=x, y=y, order=order)
    annotator.configure(test='Mann-Whitney', text_format='star', loc='inside')
    annotator.apply_and_annotate()
```

## Annotation for the stem cell markers

```python
df = pd.read_csv( "./stem_marker_scaled_df.csv", header = 0, index_col  = 0)
df.head(2)
```

```python
df.columns
```

```python
stem_marker_genes = ['Lgr5', 'Myc', 'Ascl2', 'Olfm4', 'Clu', 'Hopx', 'Tert', 'Bmi1']
```

```python
abs_treatment_color = ['#ffb482', '#a1c9f4']
order = ['WT', 'Mtgr1 -/-']


for i in range(len( stem_marker_genes)):

    f,ax = plt.subplots(figsize = [5,5])
    
    x = 'condition'
    y = stem_marker_genes[i]
    sn.violinplot(data = df,  x = x, y = y, 
               palette=abs_treatment_color, hue = 'condition',  ax = ax, )
    sn.stripplot( data = df,  x = x, y = y, 
                 dodge = True,
                   palette = ['chocolate', 'b'], hue = 'condition', size = 1.5, alpha = 0.5, ax = ax)
    ax.get_legend().remove()
    #axs.set_ylabel("gene score")
    #axs.set_title("Brush border gene expression in enterocytes")
    
    pairs=[('WT', 'Mtgr1 -/-')]
    annotator = Annotator(ax, pairs, data=df, x=x, y=y, order=order)
    annotator.configure(test='Mann-Whitney', text_format='star', loc='inside')
    annotator.apply_and_annotate()
    save_path = '~/Dropbox/MTGR1_analysis/figures/'
    plt.savefig( './figures/'+ 'stem_' + y + '.png', bbox_inches='tight' ) 
```
